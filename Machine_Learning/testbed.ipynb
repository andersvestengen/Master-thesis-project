````python
# Checking that the Dataset actually outputs the correct pairs of items, not just at the beginning. 
from Datasets.GAN_Dataset_1 import GAN_dataset
from PIL import Image
from torchvision import transforms
from torch.utils.data import DataLoader
from tqdm import tqdm
import torch
from time import time
import numpy as np
Desk_dir = "G:/Master-thesis-project/Machine_Learning"
t_dir = Desk_dir + "/output_img/Training"
v_dir = Desk_dir + "/output_img/Validation"

Settings = {
            "epochs"                : 40,
            "batch_size"            : 1,
            "L1_loss_weight"        : 100,
            "BoxSize"               : 5,
            "lr"                    : 0.0002,
            "dataset_loc"           : Desk_dir,
            "preprocess_storage"    : None,
            "seed"                  : 266, # random training seed
            "num_workers"           : 1,
            "shuffle"               : True,
            "Datahost"              : "cpu", #Should the data be located on the GPU or CPU during training?
            "Datasplit"             : 0.7,
            "device"                : "cpu",
            "ImageHW"               : 256,
            "RestoreModel"          : False,
            #No spaces in the model name, please use '_'
            "ModelName"             : "GAN_V3_Box_30",
            "Drop_incomplete_batch" : True,
            "Num_training_samples"  : None,
            "Pin_memory"            : False
            }

training_transforms = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
])

Custom_dataset = GAN_dataset(Settings, transform=training_transforms)

dataset_len = len(Custom_dataset)
print("This is dataset length:", dataset_len)

train_split = int(dataset_len*Settings["Datasplit"])
val_split = int(dataset_len - train_split)


train_set, val_set = torch.utils.data.random_split(Custom_dataset, [train_split, val_split])

train_loader = DataLoader(train_set,
                        num_workers     = Settings["num_workers"],
                        batch_size      = Settings["batch_size"], 
                        shuffle         = Settings["shuffle"],
                        drop_last       = Settings["Drop_incomplete_batch"],
                        pin_memory      = Settings["Pin_memory"])


val_loader = DataLoader(val_set,
                                num_workers = Settings["num_workers"],
                                batch_size = Settings["batch_size"], 
                                shuffle = True,
                                drop_last =Settings["Drop_incomplete_batch"],
                                pin_memory = Settings["Pin_memory"])

ToPILImageTrans = transforms.ToPILImage()

def FromTorchTraining(image):
    return image.mul(255).add_(0.5).clamp_(0,255).permute(1, 2, 0).to("cpu", torch.uint8).numpy()

def train():
    trange = tqdm(range(100), unit="images", leave=False)
    for i in trange:
        image, sample, _ = next(iter(train_loader))
        dim, dam = FromTorchTraining(image.squeeze(0)), FromTorchTraining(sample.squeeze(0))
        im = Image.fromarray(dim)
        im.save(t_dir + "/images/original" + str(i) + ".jpg") 

        sa = Image.fromarray(dam)
        sa.save(t_dir + "/samples/sample" + str(i) + ".jpg")  



def val():
    trange = tqdm(range(60), unit="images", leave=False)
    for i in trange:
        image, sample, _ = next(iter(val_loader))
        dim, dam = FromTorchTraining(image.squeeze(0)), FromTorchTraining(sample.squeeze(0))
        im = Image.fromarray(dim)
        im.save(v_dir + "/images/original" + str(i) + ".jpg") 

        sa = Image.fromarray(dam)
        sa.save(v_dir + "/samples/sample" + str(i) + ".jpg")  

train()
val()
````

````python
#testing the new Pytorch Ignite metrics
import torch
import numpy as np
from ignite.metrics import PSNR, SSIM
from skimage.metrics import peak_signal_noise_ratio as PSNR_sk
from skimage.metrics import structural_similarity as SSIM_sk
from time import time


psnr_calc = PSNR(data_range=1.0)
ssim_calc = SSIM(data_range=1.0)
N = 500
p_ssim_arr = torch.zeros(N)
p_psnr_arr = torch.zeros(N)

s_ssim_arr = torch.zeros(N)
s_psnr_arr = torch.zeros(N)

agreement_psnr  = torch.zeros(N)
agreement_ssim  = torch.zeros(N)

for i in range(N):
    image = torch.randn((1,3,256,256))
    target = image.clone() * torch.randn(1)    

    pytorch_psnr_start = time()
    psnr_calc.update((image, target))
    output_p = psnr_calc.compute()
    psnr_calc.reset()
    p_psnr_arr[i] = time() - pytorch_psnr_start # seconds 

    pytorch_ssim_start = time()
    ssim_calc.update((image, target))
    output_s = ssim_calc.compute()
    ssim_calc.reset()
    p_ssim_arr[i] = time() - pytorch_ssim_start # seconds 


    image = image.squeeze(0).permute(1,2,0).numpy()
    target = target.squeeze(0).permute(1,2,0).numpy()

    psnr_start = time()
    s_output_p = PSNR_sk(image, target, data_range=1.0)
    s_psnr_arr[i] = time() - psnr_start # seconds 

    ssim_start = time()
    s_output_s = SSIM_sk(image, target, data_range=1.0, channel_axis=2)
    s_ssim_arr[i] = time() - ssim_start # second

    agreement_psnr[i] = torch.abs(output_p - torch.tensor(s_output_p))
    agreement_ssim[i] = torch.abs(output_s - torch.tensor(s_output_s))

ssim_stop = time() - ssim_start # seconds 

print("This is the Pytorch times SSIM:", p_ssim_arr.mean(), "PSNR:", p_psnr_arr.mean())
print("This is the SKImage times SSIM:", s_ssim_arr.mean(), "PSNR:", s_psnr_arr.mean())

print("Relative difference: SSIM s/p;", s_ssim_arr.mean()/p_ssim_arr.mean(), "PSNR s/p:", s_psnr_arr.mean()/p_psnr_arr.mean())

print("Agreement: PSNR:", agreement_psnr.sum(), "SSIM:", agreement_ssim.sum())
#Change window to 7,7 and we'll probably be back to default state but with the quicker implementations. Also Check if the different losses are multiplied together or summed together.
````

````python
def ewma_vectorized(data, alpha, offset=None, dtype=None, order='C', out=None):
    """
    Calculates the exponential moving average over a vector.
    Will fail for large inputs.
    :param data: Input data
    :param alpha: scalar float in range (0,1)
        The alpha parameter for the moving average.
    :param offset: optional
        The offset for the moving average, scalar. Defaults to data[0].
    :param dtype: optional
        Data type used for calculations. Defaults to float64 unless
        data.dtype is float32, then it will use float32.
    :param order: {'C', 'F', 'A'}, optional
        Order to use when flattening the data. Defaults to 'C'.
    :param out: ndarray, or None, optional
        A location into which the result is stored. If provided, it must have
        the same shape as the input. If not provided or `None`,
        a freshly-allocated array is returned.
    """
    data = np.array(data, copy=False)

    if dtype is None:
        if data.dtype == np.float32:
            dtype = np.float32
        else:
            dtype = np.float64
    else:
        dtype = np.dtype(dtype)

    if data.ndim > 1:
        # flatten input
        data = data.reshape(-1, order)

    if out is None:
        out = np.empty_like(data, dtype=dtype)
    else:
        assert out.shape == data.shape
        assert out.dtype == dtype

    if data.size < 1:
        # empty input, return empty array
        return out

    if offset is None:
        offset = data[0]

    alpha = np.array(alpha, copy=False).astype(dtype, copy=False)

    # scaling_factors -> 0 as len(data) gets large
    # this leads to divide-by-zeros below
    scaling_factors = np.power(1. - alpha, np.arange(data.size + 1, dtype=dtype),
                               dtype=dtype)
    # create cumulative sum array
    np.multiply(data, (alpha * scaling_factors[-2]) / scaling_factors[:-1],
                dtype=dtype, out=out)
    np.cumsum(out, dtype=dtype, out=out)

    # cumsums / scaling
    out /= scaling_factors[-2::-1]

    if offset != 0:
        offset = np.array(offset, copy=False).astype(dtype, copy=False)
        # add offsets
        out += offset * scaling_factors[1:]

    return out
````

````python
#Converting a numpy vectorized EWMA for Pytorch
def SmoothCurve(input, a=0.5):
    """
    Using the exponential moving average as described here:
    https://ai.stackexchange.com/questions/17360/what-is-the-best-way-to-smoothen-out-a-loss-curve-plot
    """
    S_t = torch.zeros(input.size(0))
    S_t[0] = input[0]
    for n, Y_t in enumerate(input[1:], 1):
        S_t[n] = Y_t * (a / (1 + n)) + ( S_t[n-1]*(a / (1 + n)))
    return S_t
````

````python
import torch
import PIL 
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from Datasets.GAN_Dataset_1 import GAN_dataset
from Models.GAN_ATTN_Model import Generator_Defect_GAN
import numpy as np
from ignite.metrics import SSIM as P_SSIM
from ignite.metrics import PSNR as P_PSNR
Machine_learning_dir = "/home/anders/Master-thesis-project/Machine_Learning" # should point to the Machine learning folder of the local directory
Preprocess_dir = "/home/anders/Thesis_image_cache" # This can be set wherever you want. This is where the dataloader will create the image cache after converting the training images to tensor files.
Celeb_A_Dataset = "/home/anders/Celeb_A_Dataset"	
Standard_training_Set = "/home/anders/Master-thesis-project/Machine_Learning/Images"
#Most of these settings don't matter. They're just here to make it easier to initalize the dataloader et.al
#The fields that do matter are things like normalization, cuda, Do norm, Pin memory, preprocess storage, dataset loc, 
ssim_calc = P_SSIM(data_range=1.0, kernel_size=(7,7))
psnr_calc = P_PSNR(data_range=1.0)

def FromTorchTraining(image):
        return image.permute(1,2,0).mul_(255).clamp(0,255).to("cpu", torch.uint8).numpy()

def get_defects(image, mask):
        num_defects = torch.sum(~mask)
        side = num_defects.divide(3).sqrt().floor().to(dtype=torch.int, device="cpu")
        Whole = side**2 * 3
        if side == 8:
                real_B = torch.masked_select(image, ~mask).view(1,3,side,side)
        else:
                real_B = torch.masked_select(image, ~mask)[:-(num_defects - Whole)].view(1,3,side,side)
        return real_B

def RestoreModel(model, modeldir):
    model.load_state_dict(torch.load(modeldir, map_location=torch.device(device)))
    model.to(device)
    print("Succesfully loaded model")

Settings = {
        "epochs"                : 5,
        "batch_size"            : 1, # This must be 1 for inference!
        "L1__local_loss_weight" : 50, # Don't know how much higher than 100 is stable, 300 causes issues. Might be related to gradient calc. balooning.
        "BoxSet"               : [8,8], # min/max defect, inclusive
        "Loss_region_Box_mult"  : 1, # How many multiples of the defect box would you like the loss to account for?
        "Blockmode"             : False, #Should the defects be random artifacts or solid chunks?
        "BlackandWhite"         : False, #Whether to use black or white defects (or both)
        "Num Defects"           : 3,
        "CenterDefect"          : False, #This will disable the randomization of the defect within the image, and instead ensure the defect is always centered. Useful for initial training and prototyping.
        "lr"                    : 0.0002,
        "dataset_loc"           : Machine_learning_dir,
        "preprocess_storage"    : Preprocess_dir,
        "seed"                  : 59393, # random training seed
        "num_workers"           : 1,
        "Data_mean"             : [0.3212, 0.3858, 0.2613],
        "Data_std"              : [0.2938, 0.2827, 0.2658],
        "Do norm"               : False, #Normalization on or off 
        "shuffle"               : True,
        "Datasplit"             : 0.7,
        "device"                : "cuda",
        "ImageHW"               : 128,
        "RestoreModel"          : False,
        #No spaces in the model name, please use '_'
        "ModelTrainingName"     : "RESOURCE_TEST_DELETE_ME",
        "Drop_incomplete_batch" : True,
        "Num_training_samples"  : None, #Setting this to None makes the Dataloader use all available images.
        "Pin_memory"            : True
        }

device = "cpu"
modeldir = Machine_learning_dir + "/Trained_Models/Final_LatentInpainter_20_epochs (2023-08-31_14-34-11)/model.pt"

model = Generator_Defect_GAN(snormalization=True, dropout=True)
RestoreModel(model, modeldir)
testing_dataset_file = Machine_learning_dir + "/CELEBA_Test_split"

test_set = GAN_dataset(Settings, testing_dataset_file, preprocess=False)

imloader = DataLoader(test_set,
                                num_workers = 0,
                                batch_size = 1, 
                                shuffle = True,
                                drop_last=False)

im, test_im, mask = next(iter(imloader))
f_im, _ = model(test_im)
plt.imshow(FromTorchTraining(test_im.squeeze(0)))
plt.show()
plt.imshow(FromTorchTraining(mask.int().squeeze(0)))
plt.show()

# Now for the loss and score calc implementations 

# where(condition, input, other), return input if condition, else return other
real_B = get_defects(im, mask)
fake_B = get_defects(f_im, mask)

plt.imshow(FromTorchTraining(f_im.squeeze(0)))
plt.show()

plt.imshow(FromTorchTraining(real_B.squeeze(0)))
plt.show()

plt.imshow(FromTorchTraining(fake_B.squeeze(0)))
plt.show()

psnr_calc.update((test_im, im))
print("PSNR defect Gen value:", psnr_calc.compute())
psnr_calc.reset()
ssim_calc.update((fake_B, real_B))
print("SSIM defect Gen value:",  ssim_calc.compute())
ssim_calc.reset()

````
